```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/jorisvandenbossche/pandas-tutorial/master/data/titanic.csv')
df
```

```python
df['Pclass'].value_counts().plot.pie()
```

```python
df['Age'].mean()
```

средний возраст

```python
df['Age'].plot(kind='hist', bins=20)
```

```python
df['Age_group']=df['Age']
df['Age_group'].fillna(-1, inplace=True)
df['Age_group']= df['Age_group'].map(lambda age: int(age//10)+1)
df
```

```python
df['Age_group'].value_counts()
```

```python
age_by_survived=df.groupby(['Age_group', 'Survived'])
age_by_survived.size()
```

```python
age_by_survived=age_by_survived.size().unstack()
age_by_survived
```

```python
age_by_survived.fillna(0).plot(kind='barh', stacked=True)
```

```python
age_by_survived_withoun_na=age_by_survived[1:].fillna(0)
age_by_survived_withoun_na
```

```python
def map_to_persent(r):
  return pd.Series([float(v)/sum(r) for v in r])

age_by_survived_withoun_na.apply(map_to_persent, axis=1)
```

```python
age_by_survived_withoun_na.apply(map_to_persent, axis=1).plot(kind='barh', stacked=True)
```

```python
df.groupby(['Sex','Survived']).size().unstack().fillna(0).apply(map_to_persent, axis=1).plot(kind='barh', stacked=True)
```

```python
df.groupby(['Pclass', 'Survived']).size().unstack().fillna(0).apply(map_to_persent, axis=1).plot(kind='barh', stacked=True)
```
